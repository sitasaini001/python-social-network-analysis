```python
#Parsing the data in pure python
with open ("initialdata.txt",encoding='utf-8') as f:
    data=f.read()
    
```

```python
chunks=data.split("\n\n")
chunks=[c for c in chunks if len(c)>3]
```

```python
chunks
```

```python
chunks[0]
```

```python
def parse_chunk(chunk):
    chunk=chunk.strip()
    sep_chunk=chunk.split('\n')
    username=sep_chunk[0]
    no_Of_posts=sep_chunk[1].split(" post")[0].replace(",","")
    no_Of_followers=float(sep_chunk[2].split(" followers")[0].replace(",","").replace("K","").replace("M",""))
    if("K" in sep_chunk[2]):
        no_Of_followers=int(no_Of_followers *1000)
    elif("M" in sep_chunk[2]):
        no_Of_followers=int(no_Of_followers *1000000)    
    else:
        no_Of_followers=int(no_Of_followers)
        
    no_Of_following=float(sep_chunk[3].split("following")[0].replace(",","").replace("K","").replace("M",""))
    if("K" in sep_chunk[2]):
        no_Of_following=int(no_Of_following *1000)
    elif("M" in sep_chunk[2]):
        no_Of_followers=int(no_Of_following *1000000)
    else:
        no_Of_following=int(no_Of_following)

    name=sep_chunk[4]
    if(len(sep_chunk)>5):
       types_of_page=sep_chunk[5]
       bio="\n".join(sep_chunk[6:])
    else:
        types_of_page="unknown"
        bio=""
    return{"username":username,"no_Of_posts":no_Of_posts,  "no_Of_followers":no_Of_followers,"no_Of_following": no_Of_following,
       "name":name,"types_of_page":types_of_page,"bio":bio, }

print(parse_chunk(chunks[0]))
```

```python
all_chunks=[]
for chunk in chunks:
    parsed_chunk=parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
print(all_chunks)
```

```python
import json
s=json.dumps(all_chunks,indent=4)
with open ('data.json','w') as f:
    f.write(s)
```
